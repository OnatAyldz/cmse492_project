```python
import numpy as np
import random
import os
import matplotlib.pyplot as plt
from skimage import io
import json

def generate_eqn_image_with_boxes(folder, eqn):
    eqn_array = []
    ground_truth_boxes = []
    current_x = 0  # Keep track of x position
    
    # 1. Pick correct file and track positions
    for char in eqn:
        if char == "*": char = 'times'
        char_folder = f"{folder}/{char}"
        file = random_sample_file(char_folder)
        img = io.imread(file)
        
        # Record ground truth box
        height, width = img.shape
        ground_truth_boxes.append([current_x, 0, width, height])
        
        # Update x position for next character
        current_x += width
        
        eqn_array.append(img)
    
    # 2. Concatenate all images together
    eqn = np.hstack(eqn_array)
    return eqn, ground_truth_boxes

# Set up parameters
folder = "../data/"  # Your source data folder
output_folder = "test_equation_images_withlabels"  # Where to save generated equations
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Parameters for equation generation
EQNS = 50  # Number of test equations to generate
NUMBERS = 2  # Number of numbers in each equation
MAX_NUM = 100  # Max value for random numbers

# Modified generation loop
for i in range(EQNS):
    # Generate equation string
    eqn = equation_generator(numbers=NUMBERS, max_num=MAX_NUM)
    print(f"Generated test equation: {eqn}")
    
    # Generate image and get ground truth boxes
    eqn_array, ground_truth_boxes = generate_eqn_image_with_boxes(folder, eqn)
    filename = f"{NUMBERS}numbers_test_{str(i).zfill(3)}"
    
    # Save the image
    plt.imshow(eqn_array, cmap="gray")
    plt.axis('off')
    plt.savefig(os.path.join(output_folder, f"{filename}.png"))
    plt.close()
    
    # Save ground truth boxes to a JSON file
    ground_truth_data = {
        'equation': eqn,
        'boxes': ground_truth_boxes
    }
    with open(os.path.join(output_folder, f"{filename}_gt.json"), 'w') as f:
        json.dump(ground_truth_data, f)
```
